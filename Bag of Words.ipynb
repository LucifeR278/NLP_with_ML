```python
import pandas as pd
```

```python
messages=pd.read_csv('SMSSpamCollection.txt',sep='\t',names=["label","message"])
```

```python
messages
```

```python
## Data Cleaning and Preprocessing

import re
import nltk
nltk.download('all')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lematizer=WordNetLemmatizer()
```

```python

messages['message'][0]
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lematizer.lemmatize(word,pos='v') for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

## Create Bag of Words 

```python
from sklearn.feature_extraction.text import CountVectorizer
##  for binary BOW enable binary=True

cv=CountVectorizer(max_features=100,binary=True) # max_feature means here is that if we sort the vocabolary in descinding order and then select the top n words based on frequency

```

```python
X=cv.fit_transform(corpus).toarray()
```
